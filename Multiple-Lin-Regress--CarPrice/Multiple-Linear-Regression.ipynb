```python
df = df.replace(["gas"],0)
df = df.replace(["diesel"],1)
```

```python
df = df.replace(["convertible"],0)
df = df.replace(["hatchback"],1)
df = df.replace(["sedan"],2)
df = df.replace(["wagon"],3)
df = df.replace(["hardtop"],4)
```

```python
df.drivewheel.unique()
```

```python
df = df.replace(["rwd"],0)
df = df.replace(["fwd"],1)
df = df.replace(["4wd"],2)
```

```python
df.enginelocation.unique()
```

```python
df = df.replace(["front"],0)
df = df.replace(["rear"],1)
```

```python
engines = df.enginetype.unique()
engines[0]
```

```python
for i in range(len(engines)):
    df = df.replace(engines[i],i)
```

```python
def replacer(df,replacing_items):
    for i in range(len(replacing_items)):
        df = df.replace(replacing_items[i],i)
    return df
```

```python
df = replacer(df,engines)
```

```python
df.enginetype.unique()
```

```python
fuelsysts = df.fuelsystem.unique()
fuelsysts
```

```python
df = replacer(df,fuelsysts)
```
